```python
import pandas as pd
import numpy as np
import altair as alt
from altair import datum
```

```python
df = pd.read_pickle('amazon_conlit_goodreads_nyt.pkl')
df.head(5)
```

```python
data=df.drop(df.columns[7:12], axis=1)
data=data.drop(data.columns[8], axis=1)
data=data.drop(data.columns[9], axis=1)
data.info()
```

```python
data.dropna().shape
```

```python
final=data.drop(data.columns[11], axis=1)
final=final.dropna()
final.head()
```

```python
final=final.reset_index()
```

```python
col = ['gr_genre1', 'gr_genre2', 'gr_genre3', 'gr_genre4', 'gr_genre5', 
        'gr_genre6', 'gr_genre7', 'gr_genre8', 'gr_genre9', 'gr_genre10']
new = pd.DataFrame(final['goodreads_genres'].to_list(), columns=col)
new.head()
```

```python
for i in col:
    final[i] = new[i]
final.info()
```

```python
num_awards = []
for i in final['goodreads_awards']:
    num_awards.append(len(i))
final['goodreads_num_awards']=num_awards
```

```python
final
```

```python
genre_count = pd.DataFrame(final['gr_genre1'].value_counts())

for i in list(final.columns)[21:30]:
    genre_count = pd.concat([genre_count,final[i].value_counts()], axis=1)
    
genre_count = genre_count.fillna(0)
genre_count['number_of_books'] = genre_count.sum(axis=1)
genre_count = pd.DataFrame(genre_count['number_of_books'])
genre_count.sort_values(by=['number_of_books'], ascending=False)

genre_count = genre_count.reset_index()
genre_count.columns = ['final_genre', 'number_of_books']
genre_count
```

```python
final = final.join(genre_count, lsuffix="_left")
final
```

```python
#View 3 
click = alt.selection_multi(
    fields = ["amazon_genre"], bind = "legend")

slope_graph_amazon = alt.Chart(final, 
    title = "Slope graph of average number of Amazon reviews per year"
    ).mark_line(opacity=0.5, point=True).add_selection(click).encode(
    alt.X('amazon_year:T', title = "Year"),
    alt.Y('mean(amazon_num_reviews):Q', title = "Average number of reviews"),
    alt.Color('amazon_genre:N'),
    alt.Tooltip('amazon_genre:N'), 
    opacity = alt.condition(click, alt.value(1.0), alt.value(0.2))
).properties(
    width={"step": 100} # adjust the step parameter
)

slope_graph_amazon
```

```python
click = alt.selection_multi(
    fields = ["gr_genre1"], bind = "legend")

slope_graph_gd =  alt.Chart(final, 
    title = "Slope graph of average number of Goodreads ratings per year"
    ).mark_line(opacity=0.5, point=True).add_selection(click).encode(
    alt.X('goodreads_publish_date:T', title = "Year"),
    alt.Y('mean(goodreads_num_ratings):Q', title = "Average number of ratings"),
    alt.Color('gr_genre1:N'),
    alt.Tooltip('gr_genre1:N'), 
    opacity = alt.condition(click, alt.value(1.0), alt.value(0.2))
).properties(
    width={"step": 100} # adjust the step parameter
).transform_filter(
    (datum.gr_genre1 == "Nonfiction") | (datum.gr_genre1 == "Fiction") | (datum.gr_genre1 == "Audiobook") | (datum.gr_genre1 == "Adult") | (datum.gr_genre1 == "Biography")
)
slope_graph_gd
```

```python
# click = alt.selection_multi(
#     fields = ["amazon_genre"], bind = "legend")

boxplot_amazon = alt.Chart(final, title = "Boxplot of number of Amazon reviews"
    ).mark_boxplot().encode(
    alt.X("amazon_num_reviews:Q", title = "Number of reviews"),
    alt.Y("amazon_genre:N", title = "Genre", sort = ["Non Fiction", "Fiction"]), 
    alt.Color("amazon_genre:N"))
boxplot_amazon
```

```python
boxplot_gd = alt.Chart(final, title = "Boxplot of number of Goodreads ratings").mark_boxplot().encode(
    alt.X("goodreads_num_ratings:Q", title = "Number of ratings"),
    alt.Y("gr_genre1:N", title = "Genre", sort = ["Poetry", "Romance", "Self Help", "Economics", "Business", 
        "Cookbooks", "Picture Books", "History", "Christian", "Humor", "Horror", "Graphic Novels", "Mystery", 
        "Fantasy", "Nonfiction", "Biography", "Historical Fiction", "Fiction", "Young Adult"]), 
    alt.Color("gr_genre1:N"))
boxplot_gd
```

```python
click = alt.selection_multi(
    fields = ["amazon_genre"], bind = "legend")

scatterplot_amazon = alt.Chart(final, 
    title = "Scatterplot of number of reviews and ratings for Amazon"
    ).mark_circle(opacity = 0.6).add_selection(click).encode(
    alt.X("amazon_num_reviews:Q"), 
    alt.Y("amazon_rating:Q"), 
    alt.Color("amazon_genre:N"), 
    opacity = alt.condition(click, alt.value(1.0), alt.value(0.2))
)
    
scatterplot_amazon
```

```python
click = alt.selection_multi(
    fields = ["gr_genre1"], bind = "legend")

scatterplot_gd = alt.Chart(final, 
    title = "Scatterplot of number of reviews and ratings for Goodreads"
    ).mark_circle(opacity = 0.6).add_selection(click).encode(
    alt.X("goodreads_num_ratings:Q"), 
    alt.Y("goodreads_rating:Q"), 
    alt.Color("gr_genre1:N"), 
    opacity = alt.condition(click, alt.value(1.0), alt.value(0.2))
)
    
scatterplot_gd
```

```python
selection = alt.selection_interval(encodings = ['x'], resolve='intersect')
amazon_multiview = alt.vconcat(slope_graph_amazon.add_selection(selection), 
    boxplot_amazon.transform_filter(selection), scatterplot_amazon.transform_filter(selection))
amazon_multiview
```

```python
selection = alt.selection_interval(encodings = ['x'], resolve='intersect')
gd_multiview = alt.vconcat(slope_graph_gd.add_selection(selection), 
    boxplot_gd.transform_filter(selection), scatterplot_gd.transform_filter(selection))
gd_multiview
```

```python
view3 = alt.hconcat(amazon_multiview, gd_multiview)
view3
```
